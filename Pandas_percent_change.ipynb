```python
import numpy as np
import pandas as pd

np.random.seed(234)
aleatorios = np.random.randint(low = 950, high = 1050, size = 100)
df = pd.DataFrame(aleatorios, columns=["precio"])
df.head()
```

```python
#Líneas de posición posterior (n+1) : df.precio.shift(-1)
#Líneas en la posición actual (n) : df.precio

df['crecimiento'] = df.precio.shift(-1).div(df.precio) - 1
df.head()
```

```python
df['crecimiento_directo'] = df.precio.pct_change()
df.head()
```
